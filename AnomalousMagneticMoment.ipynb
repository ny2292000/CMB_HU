```python
import sympy

# Define the symbols
z = sympy.Symbol('z', real=True, positive=True)
A, B = sympy.symbols('A B', real=True, positive=True)

# Define the integrand
f = 1 / (z*A + (1 - z)*B)**2

# Perform the definite integral from 0 to 1
result = sympy.integrate(f, (z, 0, 1))

# Simplify the resulting expression
simplified_result = sympy.simplify(result)

print("Integral =", result)
print("Simplified =", simplified_result)
```

```python
# Observed value of a_e for normalization
a_e_observed = 1.15965218128e-3

# Contributions from QED electromagnetic terms
Schwinger_term = alpha * (1 / (2 * math.pi))  # First-order (Schwinger)
C2_contribution = alpha * alpha * (-0.328478965579193)  # Second-order
C3_contribution = alpha * alpha * alpha * (1.181241456587)  # Third-order
C4_contribution = alpha * alpha * alpha * alpha * (-1.9144)  # Fourth-order

# Cumulative sum for each order
cumulative_Schwinger = Schwinger_term
cumulative_C2 = cumulative_Schwinger + C2_contribution
cumulative_C3 = cumulative_C2 + C3_contribution
cumulative_C4 = cumulative_C3 + C4_contribution

# Normalize each contribution to the observed value
normalized_Schwinger = cumulative_Schwinger / a_e_observed
normalized_C2 = cumulative_C2 / a_e_observed
normalized_C3 = cumulative_C3 / a_e_observed
normalized_C4 = cumulative_C4 / a_e_observed

# Distance from normalized value to 1 (perfect match with observed)
distance_Schwinger = abs(1 - normalized_Schwinger)*100
distance_C2 = abs(1 - normalized_C2)*100
distance_C3 = abs(1 - normalized_C3)*100
distance_C4 = abs(1 - normalized_C4)*100

# Results
{
    "Schwinger_Normalized": normalized_Schwinger,
    "C2_Normalized": normalized_C2,
    "C3_Normalized": normalized_C3,
    "C4_Normalized": normalized_C4,
    "Distances": {
        "Schwinger": distance_Schwinger,
        "C2": distance_C2,
        "C3": distance_C3,
        "C4": distance_C4
    }
}
```
